```python
import os
import pandas as pd


def rename_admin1_gadm(iso3, path_results = '/Users/evelynm/Documents/UNU_IDMC/data/risk/CF'):
    result_paths = os.listdir(path_results+'/'+iso3)
    for result_path in result_paths:
        try:
            df_results = pd.read_csv(path_results+'/'+iso3+'/'+result_path)
            df_results.replace({'admin0':-999}, inplace=True)
            df_results['admin1'] = df_results.admin1.astype(float)
            df_results['admin1'] = df_results.admin1.astype(int)
            df_results['GID_1'] = df_results.admin1.map(gadm_attrs_map)
            df_results.to_csv(path_results+'/'+iso3+'/'+result_path)
        except:
            print(iso3)
            
```
